```python
model_data = "/content/gdrive/MyDrive/Colab_Notebooks/scriptModels/data/m-pro_07-06-2021-bin-contato.sdf"
```

```python
mols = []
y = []
for mol in Chem.SDMolSupplier(model_data):
    if mol is not None:
        mols.append(mol)
```

```python
# Calculate descriptors (fingerprints) and convert them into numpy array

# generate binary FEATMORGAN fingerprint with radius 2
fp = [AllChem.GetMorganFingerprintAsBitVect(m, 2, 2048, useFeatures=True) for m in mols]

def rdkit_numpy_convert(fp):
    output = []
    for f in fp:
        arr = np.zeros((1,))
        DataStructs.ConvertToNumpyArray(f, arr)
        output.append(arr)
    return np.asarray(output)
```

```python
# Calculate descriptors (fingerprints) and convert them into numpy array

# generate binary MORGAN fingerprint with radius 2
fp2 = [AllChem.GetMorganFingerprintAsBitVect(m, 2, 2048, useFeatures=False) for m in mols]

x2 = rdkit_numpy_convert(fp)
# Convert fingerprint to Pandas DataFram,e
#x2 = pd.DataFrame(x)
```

```python
def getNeighborsDitance(trainingSet, testInstance, k):
    neighbors_k=metrics.pairwise.pairwise_distances(trainingSet, Y=testInstance, metric='dice', n_jobs=1)
    neighbors_k.sort(0)
    similarity= 1-neighbors_k
    return similarity[k-1,:]
```

```python
distance_train_set = []
distance_test_set = []
AD = []
k= int(round(pow((len(x2)) ,1.0/3), 0))
distance_train = getNeighborsDitance(x2, x2 , k)
distance_train_set.append(distance_train)
distance_test = getNeighborsDitance(x2, x, k)
distance_test_set.append(distance_test)
Dc = np.average(distance_train_set)-(1*np.std(distance_train_set))
for i in range(len(x)):
    ad=0
    if distance_test_set[0][i] >= Dc:
        ad = 1
    AD.append(ad)
```

```python
AD = pd.DataFrame(AD)
AD = AD.rename(columns = {0:'AD'})
```

```python
pred_bin = pred_bin.join(AD)
pred_bin.head()
```
